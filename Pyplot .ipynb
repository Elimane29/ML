```python
import numpy as np
```

```python
x=np.linspace(0, 2, 10)
y= x**2
print(x)
```

```python
#1)Pyplot 

import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.show()
```

```python
#Pour un nuage de points au lieu de plot on utilise scatter 
plt.scatter(x,y)
```

```python
#Plusieurs paramètres possibles pour agir sur notre courbe 
#"c" pour la couleur, "lw" pour l'épaisseur de notre ligne, "ls" pour line style genre hachuré et autres
plt.plot(x, y, c='purple', lw= 10, ls='--')
plt.show()
```

```python
#normalement on commence par écrire plt.figure avant le plot pour avoir la feuille sur laquelle notre graphe sera tracé
#dans les parenthèses de figure on peut spécifier la taille de la figure en inch pour les axes x et y
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x, y, c='purple', lw= 10, ls='--')
```

```python
plt.figure(figsize=(3,4))
plt.plot(x, y, c='purple', lw= 10, ls='--')
```

```python
#Avec figure, on peut avoir plusieurs courbes sur la méme feuille, donner des titres, des noms aux axes  
```

```python
plt.figure()
plt.plot(x,y, label='carré') #pour la légende on doit mettre les label pour la correspondance des titres et courbes
plt.plot(x,x**3, label='cubique')
plt.title('figure 1')
plt.xlabel('Messi')
plt.ylabel('Lebron')
plt.legend()
plt.show() #pour afficher la courbe mais pas indispensable pour jupyter
```

```python
#on a aussi la fonction subplot pour faicliter le fait d'avoir plusieurs graphes sur une figure

plt.subplot(2, 1, 1)
plt.plot(x, y, c='red')
plt.subplot(2, 1, 2)
plt.plot(x, y, c='blue')
```
