```python
#C'est un dataset qui contient 150 fleurs d'iris réparties en 3 classes. Les classes sont définies en 4 variables:
#longueur et largeur du pétale de la fleur et longueur et largeur du sépal de la fleur

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
%matplotlib inline


x=iris.data #tableau où on a nos variables 
y=iris.target #tableau où on a nos différentes classes
names=list(iris.target_names)

print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables') #ici quand on dit 0 et 1 c'est axis 0 et axis 1
print (f'il y a {np.unique(y).size} classes')
x #avec nos variables 150 lignes et 4 colonnes
```

```python
y #avec nos 3 classes 
```

```python
#Pour nous visualiser tout ça, plt scatter avec en abcisses 2 des 4 variables de notre problème, en mettant c=y on différencie
#les points selon leurs classes mais que selon les 2 seules variables qu'on a achoisi d'où la petite imprécision il choisit les couleurs lui-méme 

plt.scatter(x[:,0] , x[:,1], c=y)
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
plt.show()
```

**Pour visualiser des surfaces sur des graphiques 3D**

```python
f = lambda x,y: np.sin(x) + np.cos(x+y) #fonction de x et y f(x,y); lambda générateur de fonctions
```

```python

#pour visualiser ce graphe en 3D il faudra créer 2 vecteurs numpy pour les axes x et y qui vont de 0 à 5 avec la fonction meshgrid 
#pour avoir une grille de 2 axes

X=np.linspace(0, 5, 100)
Y=np.linspace(0, 5, 100)
X,Y= np.meshgrid(X, Y)

#On va injecter X et Y dans la fonction f pour que x et y aient des valeurs 
Z= f(X, Y)
```

```python
#On vérifie les dimensions de z
Z.shape
```

```python
#pour visualiser tout ça
ax=plt.axes(projection='3d')
ax.plot_surface(X, Y, Z,cmap='plasma') #cmap c'est la couleur de notre surface
plt.show()
```

**HISTOGRAMME**

```python
#L'histogramme permet de voir la distribution des données avec lesquelles on travaille, si la distribution est symétrique 
#asymétrique, d'avoir une idée de la moyenne
#on peut appliquer cela au problème d'iris

plt.hist(x[:,0]) #sur l'une de nos variables 
plt.show()
```

```python
#en abcisse on a la longueur du sépal de notre iris, le plus petit c'est 4,5 et le plus long c'est 8 cm ou inch jsplus
#en gros on a 25 fleurs qui ont une longueur de sépal autour de 5,5
```

```python
#le paramètre bins sert à compter le nombre de bandes qui par défaut est à 10 comme ci-dessus
#si on augmente ce sera plus précis

plt.hist(x[:,0], bins=30) 
plt.show()

#On voit bien que c'est plus précis
```

```python
#Possible de faire 2 histogrammes sur une meme figure pour 2 variables différentes 
plt.hist(x[:,0], bins=30)
plt.hist(x[:,1], bins=30)
plt.show()
```

```python
#on peut aussi tracer des histogrammes en 2D pour voir la distribution

plt.hist2d(x[:,0], x[:,1], cmap='Blues' )
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
plt.colorbar() #pour bien comprendre la distribution avec une légende
plt.show()
```

**Contour Plot**

```python
#Ces graphiques sont utiles pour visualiser un modèle en 3d vue du dessus
plt.contour(X,Y,Z, 20, colors='blue')
plt.show()
```

```python
#pour les lignes continues c'est positif et pour les lignes négatives c'est discontinu
```

```python
plt.contourf(X,Y,Z, 30, cmap='RdGy')
plt.colorbar()
plt.show()
```
